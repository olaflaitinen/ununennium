```python
# Normalization methods
data = torch.randn(4, 256, 256) * 1000 + 500
minmax = (data - data.min()) / (data.max() - data.min())
zscore = (data - data.mean()) / data.std()
print(f'MinMax range: [{minmax.min():.3f}, {minmax.max():.3f}]')
```
