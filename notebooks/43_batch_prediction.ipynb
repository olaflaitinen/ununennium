## Core Workflow

```python
import torch
from ununennium.models import create_model
```

```python
model = create_model('unet', in_channels=4, num_classes=5, backbone='resnet18', pretrained=False)
model.eval()
batch_size = 8
with torch.no_grad():
    for i in range(5):
        batch = torch.randn(batch_size, 4, 256, 256)
        out = model(batch)
        print(f'Batch {i+1}: {out.shape}')
```
